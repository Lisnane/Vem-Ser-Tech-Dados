1) Qual foi o time que teve a maior sequencia de vitórias durante toda a temporada?

```python
query_1 = """
SELECT
    time_serie_a AS Time,
    Vitorias_consecutivas AS Maior_Sequencia_Vitorias
FROM
    projeto_bd.classificacao_2023
ORDER BY
    Vitorias_consecutivas DESC
LIMIT 1;
"""
pd.read_sql_query(query_1, con = conn)

```

2) Qual foi o time que teve a maior sequencia de derrotas durante toda a temporada?

```python
query_2 = """
SELECT
    time_serie_a AS Time,
    derrotas_consecutivas AS Maior_Sequencia_Derrotas
FROM
    projeto_bd.classificacao_2023
ORDER BY
    Derrotas_consecutivas DESC
LIMIT 1;
"""
pd.read_sql_query(query_2, con = conn)

```

3) Qual time teve a maior sequencia de jogos marcando gols e em qual rodada ele atingiu esta marca?

```python
query_3 = """
SELECT
    time_serie_a AS time,
    Rodada,
    jogos_marcando AS Maior_Seq_Marcando
FROM
    projeto_bd.classificacao_2023
ORDER BY
    jogos_marcando DESC
LIMIT 1;
"""
pd.read_sql_query(query_3, con=conn)
```

4) Qual turno do campeonato teve a maior quantidade de gols, sabendo que o primeiro turno é da rodada 1 a 19 e o segundo turno é da rodada 20 a 38?

```python
query_4 = """
WITH Turno_Gols AS (
    SELECT
        temporada,
        rodada,
        SUM(mandante_placar + visitante_Placar) AS Total_Gols
    FROM
        projeto_bd.rodadas_2023
    GROUP BY
        temporada,
        rodada
)
SELECT
    temporada,
    CASE
        WHEN rodada BETWEEN 1 AND 19 THEN 'Primeiro Turno'
        WHEN rodada BETWEEN 20 AND 38 THEN 'Segundo Turno'
    END AS Turno,
    SUM(Total_Gols) AS Gols_Marcados
FROM
    Turno_Gols
GROUP BY
    temporada,
    Turno
ORDER BY
    Gols_Marcados DESC
LIMIT 1;
"""

pd.read_sql_query(query_4, con = conn)
```

5) Quantos times experimentaram a zona de rebaixamento na temporada?

```python
query_5 = """
SELECT 
    COUNT(time_serie_a) AS Contagem_Times_Z4
FROM (
    SELECT 
        time_serie_a
    FROM 
        projeto_bd.classificacao_2023
    WHERE 
        posicao >= (SELECT MAX(Posicao) - 3 FROM projeto_bd.classificacao_2023)
    GROUP BY 
		time_serie_a
) AS contagem;
"""

pd.read_sql_query(query_5, con = conn)
```

5.1) Quais foram esses times e por quantas rodadas eles estiveram no Z4?

```python
query_6 = """
WITH Ultimas_Posicoes AS (
    SELECT 
		time_serie_a, 
		posicao
    FROM 
		projeto_bd.classificacao_2023
    WHERE 
		posicao >= (SELECT MAX(posicao) - 3 FROM projeto_bd.classificacao_2023)
)
SELECT 
	time_serie_a, 
	COUNT(*) AS Rodadas_no_Z4
FROM 
	Ultimas_Posicoes
GROUP BY 
	time_serie_a
ORDER BY 
	Rodadas_no_Z4 DESC;
"""

pd.read_sql_query(query_6, con = conn)
```

6) E quantos times distintos experimentaram o G4?

6.1) Quais foram esses times e por quantas rodadas eles estiveram no G4?

7) Quais foram os 5 times mais vitoriosos da temporada? Algum deles venceu o campeonato?

```python
query_9 = """
SELECT
    time_serie_a,
    MAX(vitoria) AS Total_Vitorias,
    MAX(CASE WHEN Rodada = 38 AND Posicao = 1 THEN 'Sim' ELSE 'Não' END) AS Campeao
FROM
    projeto_bd.classificacao_2023
GROUP BY
    time_serie_a
ORDER BY
    Total_Vitorias DESC
LIMIT 5;
"""

pd.read_sql_query(query_9, con = conn)
```

8) Se o brasileirão fosse um time, em que posição ele estaria na ultima rodada (38), considerando a média de vitórias, derrotas e empates?
Sabendo que cada vitória vale 3 pontos, empate 1 e derrotas 0.

```python
query_10 = """
SELECT 
	COUNT(*) + 1 AS Posicao_Campeonato
FROM (
    SELECT 
		time_serie_a, 
		SUM(vitoria * 3 + empate) AS Pontos
    FROM 
		projeto_bd.classificacao_2023
    WHERE 
		rodada = 38
    GROUP BY 
		time_serie_a
) AS Tabela
WHERE Pontos > (
    SELECT 
		AVG(vitoria * 3 + empate) AS Pontos_Campeonato
    FROM 
		projeto_bd.classificacao_2023
    WHERE 
		Rodada = 38
);
"""

pd.read_sql_query(query_10, con = conn)
```

9) Quais foram as 3 melhores defesas da temporada?

11) Qual foi a média de gols dentro e fora de casa a cada rodada do Brasileirão?

```python
query_15 = """
SELECT
    rodada,
    AVG(mandante_Placar) AS media_gols_em_casa,
    AVG(visitante_Placar) AS media_gols_fora_casa
FROM
    projeto_bd.rodadas_2023
GROUP BY
    rodada
ORDER BY
    rodada;
"""

pd.read_sql_query(query_15, con = conn)
```
